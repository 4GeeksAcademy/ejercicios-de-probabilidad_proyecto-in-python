## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

# Función del lanzamiento de dados:
def lanzar_dados():
    dado_1 = random.randint(1, 6)
    dado_2 = random.randint(1, 6)
    return dado_1 + dado_2

# Número de lanzamientos:
num_lanzamientos = 1000

# Contadores para mantener un registro de las sumas mayores a 7 y pares
sumas_mayor_que_7_o_par = 0

# Realiza la simulación
for i in range(num_lanzamientos):
    suma = lanzar_dados()
    if suma > 7 or suma % 2 == 0:
        sumas_mayor_que_7_o_par += 1

# Calculamos la probabilidad 
probabilidad = sumas_mayor_que_7_o_par / num_lanzamientos

print(f"La probabilidad de sacar una suma mayor a 7 o un número par es:", {probabilidad})
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
Caja_bolas = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        Caja_bolas[i] = "blancas"
    elif (i > 9) and (i < 30):
        Caja_bolas[i] = "rojas"
    else:
        Caja_bolas[i] = "verdes"

print(Caja_bolas)

import numpy as np

def coger_bolas(iteraciones = 1000):
    count_1 = 0
    count_2 = 0

    for i in range(iteraciones):
        colores = []

        # Sacamos 5 bolas de la caja
        for i in range(5):
            colores.append(Caja_bolas[np.random.randint(0, 59)])

        # La convertimos en un array
    
        colores = np.array(colores)
        
        Suma_blancas = sum(colores == "blancas")
        Suma_rojas = sum(colores == "rojas")
        Suma_verdes = sum(colores == "verdes")
        # Añadimos al conteo:
        if (Suma_blancas == 3) and (Suma_rojas == 2):
            count_1 += 1
        
        if (Suma_blancas == 5) or (Suma_rojas == 5) or (Suma_verdes == 5):
            count_2 += 1
    
    return count_1 / iteraciones, count_2 / iteraciones

probabilidades = coger_bolas(1000)

print(f"La probabilidad de salir 3 blancas y 2 rojas es: {np.round(probabilidades[0] * 100, 2)}%")
print(f"La probabilidad de salir todas del mismo color es: {np.round(probabilidades[1] * 100, 2)}%")

 
   
 
```
